Oftmals kommt es vor, dass du ein bestehendes `DataFrame` hast, und auf Grundlage dieses `DataFrames` ein neues erzeugen willst. Dabei ist es gut zu wissen, dass bestehende Zeilen einfach ausgewählt werden können, um diese ganz einfach an das neue `DataFrame` anhängen zu können.
<br>

Hierbei kommt die Funktion **`.append()`** zum Einsatz.

![title](./pic/zuweisen/existierende_zeile/2_append.png)

```python
df2 = pd.DataFrame()
```

Zum Beispiel soll dem neuen `DataFrame` der 7. Eintrag eines anderen `DataFrames` angefügt werden. Dies würde man wie folt umsetzen:

```python
df2 = df2.append(df.loc[7])
df2
```

Auch innerhalb einer Schleife kann `.append()` verwendet werden...

```python
df2 = pd.DataFrame()

for i in range(len(df)):
    if i % 10 == 0:
        df2 = df2.append(df.iloc[i], sort=False)
```

```python
df2.head()
```

![title](./pic/zuweisen/existierende_zeile/3_append_condition.png)

...oder auf Grundlage von bestimmten Featureeigenschaften

```python
df3 = pd.DataFrame()

df3.append(df.loc[df['Age'] < 10])
```

Beachte: Dem neuen `DataFrame` wird der ursprüngliche Index des `DataFrames` übergeben, den die Zeile auch in dem alten `DataFrame` hatte. Willst du dass der Index vom neuen `DataFrame` wieder bei 0 beginnt, verwende das Argument `ignore_index=True`

```python
df4 = pd.DataFrame()

df4.append(df.loc[df['Age'] < 10], ignore_index=True)
```
