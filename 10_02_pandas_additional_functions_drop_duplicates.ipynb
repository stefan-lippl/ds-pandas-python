## Zusätzliche Attribute für `.drop_duplicates()`

### Mit dem Attribut `inplace=`

```python
df3 = df.copy()
```

Mit dem Attribut `inplace` kannst du die Ausgabe der Funktion `.drop_duplicates` auf stumm schalten. Du bekommst also bei `inplace=True` kein neues `DataFrame` ausgegeben, sondern musst dies seperat aufrufen

```python
df3.drop_duplicates(subset='PassengerId', inplace=True)
```

```python
df3
```

---

### Mit dem Attribut `ignore_index=`

```python
df4 = df.copy()
```

Wie auch bei den letzten Funktionen, setzt das Attribut `ignore_index=True` deinen Index zurück und beginnt eine aufsteigende Nummerierung mit `0` bis hin zur `Länge-1` deines `DataFrames`. Die `default`-Einstellung von diesem Attribut ist auf `False` gesetzt. Du musst also explizit `True` angeben, wenn du willst dass der Index zurückgesetzt wird.

```python
df4.drop_duplicates(ignore_index=True)
```
