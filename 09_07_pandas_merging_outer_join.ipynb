## Merge function: `outer_join` mit `left_on=` & `right_on=`

![title](./pic/merging/outer_join/5_outer_olor.png)

```python
df_passenger.rename({'PassengerId': 'pass_id'}, axis=1, inplace=True)
df_passenger.head()
```

```python
pd.merge(df_ids,
         df_passenger,
         how='outer',
         left_on='PassengerId',
         right_on='pass_id'
         )
```
