## Merge function: `inner_join` mit `on=`

![title](./pic/merging/column_wise/2_column.png)

```python
df_passenger.head()
```

```python
df_ids.head()
```

```python
pd.merge(df_ids[['PassengerId']],
         df_passenger,
         how='inner',
         on='PassengerId'
         )
```

## Merge function: `inner_join` mit `left_on=` & `right_on=`

```python
df_passenger.rename({'PassengerId': 'pass_id'}, axis=1, inplace=True)
df_passenger.head()
```

```python
pd.merge(df_ids[['PassengerId', 'Survived']],
         df_passenger[['pass_id', 'Name']],
         how='inner',
         left_on='PassengerId',
         right_on='pass_id',
         )
```
