```python
import pandas as pd
```

```python
df = pd.read_csv('./csv/titanic.csv')
df.head()
```

<br>

`df.pop()` ist `df.drop()` sehr ähnlich, aber `.drop()` gibt das `DataFrame` zurück, aus dem die Spalte entfernt wurde. Wohingegen `.pop()` die Spalte selbst als neue `pd.Series` zurückgibt.

Ich verwende `.pop`, wenn ich eine Dummy-Spalte erstellen will. Eine Spalte, die nur vorübergehend verwendet wird, und ich nicht will, dass mein `pd.DataFrame` schlampig wird.

Wenn du mehrere Spalten aus deinem Datensatz entfernen musst, kannst du entweder `.pop()` mehrmals ausführen oder stattdessen pandas `.drop()` verwenden. Mit `.drop()` kannst du eine Liste von Spalten (oder Zeilen) übergeben, die dann aus deinem `DataFrame` entfernt werden.

**Pop** nimmt keine zusätzlichen Argumente entgegen, so dass du dir keine Gedanken über Änderungen machen musst. 

<br>

### Beispiel 1: Die Spalte `Age` extrahieren

```python
df_age = df.pop('Age')
```

### Beispiel 2: Die Spalte `Sex` extrahieren

```python
df_sex = df.pop('Sex')
```

```python
df_sex
```

```python
df
```

Wie du nun also sehen kannst, wurden mithilfe von `.pop()` komplette Spalten aus dem `DataFrame` entfernt und bilden nun alleinstehend neue `Series`. Auf diesen Series kann z.B. gearbeitet und Funktionen angewandt werden um sie später wieder ans `DataFrame` anhängen zu können.
