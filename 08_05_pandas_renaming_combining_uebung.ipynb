### 5) (S) Szenario: 
#### Aktuell benutzt unsere Marketingabteilung eine Datei namens `df_cars` welche lediglich Daten zum Auto beinhaltet wie Marke und Modell. Die Abteilung Einkauf benutzt eine andere Datei (`df_car_details`)um Preise unterschiedlicher Autos zu vergleichen, wobei hier lediglich die KM Anzahl eine Rolle spielt. Um genauere Analysen machen zu können, müssen wir die beiden `DataFrames` miteinander kombinieren. Die Gemeinsamkeit beider `DataFrames` ist die Spalte `car_id`. Diese beinhaltet immer einen genauen Identifizierungsschlüssel und symbolisiert in jedem der beiden `DataFrames` ein und das selbe Auto. Deine Aufgabe ist es einen Weg zu finden, die beiden `DataFrames` miteiander zu verbinden um ein großes `DataFrame` namens `df_joined` zu erhalten. Beachte, dass das `DataFrame` `df_car_details` beschädigt wurde und nicht mehr über alle Zeilen verfügt.

```python
df_joined = df_cars.join(df_car_details.set_index('car_id'), on="car_id")
```
