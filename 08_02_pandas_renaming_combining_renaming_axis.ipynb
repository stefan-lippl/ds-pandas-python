Oft kommen Daten mit Spaltennamen, Indexnamen oder anderen Namenskonventionen zu uns, mit denen wir nicht zufrieden sind. In diesem Fall lernst du, wie du mit Hilfe von `Pandas`-Funktionen die Namen der fehlerhaften Einträge in etwas Besseres ändern kannst.

Du wirst auch lernen, wie du Daten aus mehreren `DataFrames` und/oder `Serien` kombinieren kannst.

---

## Axen umbenennen

Du wirst wahrscheinlich sehr oft Spalten umbenennen, aber sehr selten Indexwerte. Dafür ist `set_index()` normalerweise besser geeignet.

Sowohl der Zeilenindex als auch der Spaltenindex können ein eigenes Namensattribut haben. Die ergänzende Methode `rename_axis()` kann verwendet werden, um diese Namen zu ändern. Zum Beispiel:

### Rows

```python
df.rename_axis("passenger_ids", axis='rows')
```

---

### Columns

```python
df.rename_axis("passenger_dates", axis='columns')
```

---

### Rows & Columns

```python
df.rename_axis("passenger_ids", axis='rows').rename_axis("passenger_dates", axis='columns')
```
