### 2) (M) Merge die beiden `DataFrames` (`df_result1`), sodass das resultierende `DataFrame` lediglich die übereinstimmenden Zeilen beider `DataFrames` beinhaltet
##### Identische Spalte in beiden: car_id

### 6) (S): Extrahiere alle Autos der Marke `Volkswagen` aus dem `DataFrame` `df_car_basics` und speicher diese in ein neues `DataFrame` namens `df_volkswagen` wobei Zeilen mit `Nan` Values direkt gelöscht werden sollen. Merge `df_volkswagen` mit dem `DataFrame` `df_car_infos` zu einem neuen `DataFrame` namens `df_merged_volkswagen`. 

### <u>Für wieviele Fahrzeuge **liegen keine weiteren Informationen** vor?</u>

```python
df_volkswagen = df_car_basics.query('marke == "Volkswagen"').dropna()
df_volkswagen
```

```python
df_merged_volkswagen = pd.merge(df_volkswagen,
                                df_car_infos,
                                how='left',
                                left_on='car_id',
                                right_on='unique_car_id')
```

```python
df_merged_volkswagen
```

```python
len(df_merged_volkswagen[df_merged_volkswagen.unique_car_id.isna()])
```
