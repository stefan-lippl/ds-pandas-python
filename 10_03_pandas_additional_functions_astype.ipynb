Die Methode `.astype()` wird verwendet, um ein `Pandas`-Objekt in einen bestimmten **Datentyp umzuwandeln**. Die Funktion `.astype()` bietet auch die Möglichkeit, jede geeignete vorhandene Spalte in einen kategorischen Typ umzuwandeln.

`.astype()` ist sehr nützlich, wenn du einen bestimmten **Spaltendatentyp in einen anderen Datentyp umwandeln** willst. Darüber hinaus kannst du auch ein Python-`Dictionary` verwenden, um mehr als einen Spaltentyp auf einmal zu ändern. Die Schlüsselbezeichnung im `Dict` entspricht dem Spaltennamen und die Wertebezeichnung im Dictionary entspricht den neuen Datentypen, die die Spalten haben sollen.

Da `.astype()` nicht mit `Nan` Werten umgehen kann, müssen wir diese zuerst ausfindig machen und anschließend aus dem zu `DataFrame` entfernen.
Um die Daten jedoch nicht zu verlieren, speichern wir diese in ein seperates `DataFrame` ab.

```python
df_age_null = df[df['Age'].isnull()]
df_age_null
```

...Anschließend droppen wir die Zeilen, welche in der Spalte Alter `Nan` Werte haben...

```python
df = df.dropna(subset=['Age'])
df
```

...und können nun endlich den cast von `float` zu `int` vollziehen

```python
df['Age'] = df['Age'].astype(int)
df
```

Abschließend konkatenieren wir die beiden `DataFrames` wieder zu einem finalen `DataFrame`

```python
df_final = pd.concat([df, df_age_null], ignore_index=False)
df_final
```

-> `.concat` versucht die Daten einer Spalte immer dem selben Datentypen zuzuweißen. Da `Nan` vom Datentyp `float` ist, werden auch die `int` Werte wieder zu `float` gecastet
