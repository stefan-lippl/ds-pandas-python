<br>

Pandas `.startswith()` ist eine weitere Methode zum **Suchen und Filtern** von Textdaten in `Series` oder `DataFrames`. Diese Methode ist der `Python`-Methode `startswith()` ähnlich, hat aber andere Parameter und funktioniert nur mit `Pandas`-Objekten. Daher muss vor dem Aufruf dieser Methode immer `.str` vorangestellt werden, damit der `Compiler` weiß, dass es sich um eine andere als die Standardfunktion handelt.


![title](./pic/aditional_functions/startswith/2_startswith.png)

Wollen wir z.B. alle Passagiere, deren Namen mit "K" starten, können wir das wie folgt abfragen:

```python
df['Name'].str.startswith('K')
```

Die Methode `.startswith()` gibt eine `Bool`-`Series` zurück. Mit dem `[]`-Operator können wir uns die kompletten Einträge des `DataFrames` anzeigen lassen.

```python
df[df.Name.str.startswith('Ke', na=False)]
```
