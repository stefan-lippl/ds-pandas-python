### 3) (M) Erstelle eine `neue Spalte` namens `familiengerecht`, die die Werte "ja" oder "nein" beinhaltet, je nachdem ob die Anzahl der Sitzplätze <= 2 oder >2 ist

```python
for i in range(len(df)):
    anz_sitz = df.loc[i, 'anzahlSitzplaetze']
    
    if anz_sitz > 2:
        df.loc[i, 'familiengerecht'] = "ja"
    else:
        df.loc[i, 'familiengerecht'] = "nein"
```

### 4) (S) Erstell ein **neues leeres** `DataFrame`. **Füge** diesem dann alle Autos der `Marke BMW` hinzu (append).

#### 4.1 Schlechte Variante mit `for`-loop

```python
df_bmw_a = pd.DataFrame()
```

```python
for i in range(len(df)):
    if df.loc[i, 'marke'] == 'BMW':
        df_bmw_a = df_bmw_a.append(df.loc[i])
```

```python
df_bmw_a.head()
```

```python
df_bmw_a.shape
```

#### 4.2 Gute Variante mit `Pandas`

```python
df_bmw_b = pd.DataFrame()
```

```python
df_bmw_b = df_bmw_b.append(df.loc[df['marke'] == "BMW"], ignore_index=True)
```

### 7) <u>(S)</u> Lass dir alle `Porsche` & `BMW` ausgeben, deren Kilometerstand `Nan` ist und speicher diese in ein neues `DataFrame` namens `df_km_na`

```python
df_km_na = df[(df['marke'] == "Porsche") & (df.kilometer.isnull()) | 
              (df['marke'] == "BMW") & (df.kilometer.isnull())]
```

```python
df_km_na.index
```
