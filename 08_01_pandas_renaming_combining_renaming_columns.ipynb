## Renaming Columns

Die erste Funktion, die wir hier vorstellen, ist `rename()`, mit der du Indexnamen und/oder Spaltennamen ändern kannst. Um zum Beispiel die Spalte **"PassengerId"** in unserem Datensatz in **"pass_id"** umzubenennen, würden wir Folgendes tun:

```python
df.rename(columns={'PassengerId': 'pass_id'})
```

<br>

`.rename()` ermöglicht die Umbenennung von Index- oder Spaltenwerten durch Angabe eines Index- bzw. Spaltenschlüsselwortparameters. Die Funktion unterstützt eine Vielzahl von Eingabeformaten, aber normalerweise ist ein Python-`Dictionary` am praktischsten. Hier ein Beispiel für die Umbenennung einiger Elemente des Indexes.

```python
df.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```
